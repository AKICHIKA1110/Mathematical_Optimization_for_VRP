```python
#ライブラリ＆APIを呼び出し
import pulp

#SLEは任意のname，最大化問題を定義
problem = pulp.LpProblem("SLE", pulp.LpMaximize)

#変数を定義 cat='Continuous'は連続実数を意味する
x = pulp.LpVariable("x", cat='Continuous')
y = pulp.LpVariable("y", cat='Continuous')

#制約式の定義
problem += 120 * x + 150 * y == 1440
problem += x + y == 10

status = problem.solve()

print(f"Status: {pulp.LpStatus[status]}")  
print(f"x = {x.varValue}")
print(f"y = {y.varValue}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データの作成
x = np.linspace(0, 10, 100)
y1 = (1440 - 120 * x) / 150  # 120x + 150y = 1440
y2 = 10 - x                 # x + y = 10

# グラフの描画
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label="120x + 150y = 1440")
plt.plot(x, y2, label="x + y = 10")
plt.plot(2, 8, 'ro') # 交点 (2, 8) を赤丸で表示
plt.xlabel("Apples (x)")
plt.ylabel("Oranges (y)")
plt.legend()
plt.grid(True)

# 画像として保存（LaTeXファイルと同じフォルダに保存してください）
plt.savefig("graph2.1.png")
```
